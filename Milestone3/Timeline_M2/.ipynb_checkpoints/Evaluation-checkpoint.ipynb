```python
import onnx
from Predictor import Predictor
import torch
import os
import torchvision.models as models
import numpy as np
import torch.nn as nn

```

```python
# MLP for classification

model = models.efficientnet_b7(weights="IMAGENET1K_V1")

#resnet not trained later 
for param in model.parameters():
    param.requires_grad = False

num_features = model.classifier[-1].in_features
mlp = nn.Sequential(
    nn.Linear(num_features, 512),
    nn.ReLU(),
    nn.Dropout(p=0.8),
    nn.Linear(512, 64),
    nn.ReLU(),
    nn.Dropout(p=0.8),
    nn.Linear(64, 6),
    nn.Softmax(dim=1)
)

model.classifier = mlp
model_path = "/home/mdelabrassinne/DSproject/Milestone2/Frames Classification/Models/efficientNetB7/model_efficientNetB7.pt"

model.load_state_dict(torch.load(model_path, map_location=torch.device('cpu')))



# Put the model in evaluation mode
model.eval()

# Create an example input tensor
input_tensor = torch.ones(1, 3, 224, 224)

# Use the model to make a prediction
with torch.no_grad():
    output_tensor = model(input_tensor)

# Convert the output tensor to a numpy array
output_array = output_tensor.numpy()

# Print the prediction
print(output_array)
print(np.sum(output_array))
```
